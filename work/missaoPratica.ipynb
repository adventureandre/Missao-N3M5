```python
import pandas as pd
import numpy as np

dados = pd.read_csv('dadosMissao.csv', sep=';', engine='python', encoding='utf-8')

print("Informações gerais:")
print(dados.info())

print("\nPrimeiras linhas do conjunto:")
print(dados.head())

print("\nÚltimas linhas do conjunto:")
print(dados.tail())

dados2 = dados

dados2['Calories'].fillna(0, inplace=True)
print("\nRemovendo valores nulos no calories:")
print(dados2)

dados2['Date'].fillna('1900/01/01', inplace=True)
print("\nsubstituir valores nulos em 'Date':")
print(dados2)


# aki muda o campo Date para Datatime
dados2['Date'] = pd.to_datetime(dados2['Date'])


# aki volta pra NaN
try:
    dados2['Date'] = pd.to_datetime(dados2['Date'])
except Exception as e:
    print("Erro ao transformar 'Date' em datetime:", e)



dados2['Date'] = pd.to_datetime(dados2['Date'])
print("\n transformação da coluna 'Date' em datetime:")
print(dados2)


dados2['Date'] = dados2['Date'].replace('20201226', '2020-12-26')
dados2['Date'] = pd.to_datetime(dados2['Date'], errors='coerce')

print("\n substituir '20201226':")
print(dados2)

# Passo 11: Remover registros contendo valores nulos
dados_final = dados2.dropna()
print("\nApós remover registros com valores nulos:")
print(dados_final)

# Exibir as informações finais
print("\nInformações gerais do DataFrame final:")
print(dados_final.info())
```
